```python
from python_bugreport_parser.bugreport.bugreport_txt import BugreportTxt
from python_bugreport_parser.plugins import PluginRepo
from glob import glob


def analyze(row):
    error_time = row[4].to_pydatetime()
    # print(type(error_time))
    path = f"/home/dayong/jira/o3-feedback/{row[0]}"
    if not os.path.exists(path):
        return

    bugreport_txt_paths = glob(str(Path(path) / "**/bugreport*.txt"))
    if bugreport_txt_paths:
        bugreport_txt_path = bugreport_txt_paths[0]
    else:
        return
    print(bugreport_txt_path)
    bugreport_txt = BugreportTxt(bugreport_txt_path)
    bugreport_txt.load()
    bugreport_txt.set_error_timestamp(error_time)

    plugin = PluginRepo.find_by_name("InvalidBugreportPlugin")
    if bugreport_txt.loaded:
        plugin.analyze(bugreport_txt)
        print(plugin.report())


df.apply(analyze, axis=1)
```
